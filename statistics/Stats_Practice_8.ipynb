```python
row1 = np.array([40,45,25,10])
row2 = np.array([35,30,20,30])
```

```python
sum_r1 = np.sum(row1)
sum_r2 = np.sum(row2)
sum_row = np.array([sum_r1, sum_r2])
sum_row
```

```python
sum_cal = row1 + row2
sum_cal
```

```python
exp = []
for i in sum_row:
  for j in sum_cal:
    value = (i*j)/235
    exp.append(value)
exp
```

```python
obj = np.array([40,45,25,10,35,30,20,30])
obj
```

```python
np.sum((np.square(obj - exp))/exp)
```

### As we have taken,

### H0 = Not Associated

### Ha = Associated

### And we have 13.788 as the chi calculated value which is greater than the chi table value i.e 7.815.
### We can say that the observations are associated.
