```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Central Limit Theorem

```python
pop_data = [np.random.randint(10, 100) for i in range(1000)]
pop_table = pd.DataFrame(pop_data, columns=['Pop_Data'])
pop_table
```

```python
np.mean(pop_data)
```

```python
plt.figure(figsize=(4,3))
sns.kdeplot(x = "Pop_Data", data = pop_table)
plt.show()
```

```python
sample_mean = []
for no_sample in range(50):
  sample_data = []
  for data in range(500):
    sample_data.append(np.random.choice(pop_data))
  sample_mean.append(np.mean(sample_data))
```

```python
# Almost the same mean as overall population data(pop_data)
np.mean(sample_mean)
```

```python
sample_M = pd.DataFrame(sample_mean, columns=['Sample_Mean'])
sample_M
```

```python
# Likely more normally distributed as compared to the overall population data's distribution
plt.figure(figsize=(4,3))
sns.kdeplot(x = "Sample_Mean", data = sample_M)
plt.show()
```
